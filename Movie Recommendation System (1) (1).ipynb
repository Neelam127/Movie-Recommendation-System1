```python
col_names = ['user_id', 'item_id', 'rating', 'timestamp']  #giving column names for data set

df = pd.read_csv('C:\\Users\\Neelam\\Documents\\Movie-Recommendation-System1-main (2)\\Movie-Recommendation-System1-main\\u.data', sep='\t',names=col_names)  #reading data set file

#sep='\t' beacuse u.data set is tab seprated contents and names is to assign column names from the list created

df.head()  #displays the first 5 data set
```

## Get Movie Titles Data

```python
movie_titles = pd.read_csv('C:\\Users\\Neelam\\Documents\\Movie-Recommendation-System1-main (2)\\Movie-Recommendation-System1-main\\Movie_ID_Titles')  #Reading Movie Titles data set file which is w.r.t item_id

movie_titles.head()    #displaying 1st 5 data set
```

## Merging User Data and Movie Title Data Table

### merging key will be the item ID column

```python
df = pd.merge(df,movie_titles,on='item_id')   #using pandas merge function for merging two tables

df.head()
```

# EDA (Exploratory Data Analysis)

## Creating ratings dataframe with average rating and number of ratings

```python
df.groupby('title')['rating'].mean().sort_values(ascending = False).head()  

#first we will groupby operator for title and rating
#then by using mean function we will take average
#then we will use sort function and make ascending=false so that it will display from Highest to lowest
```

```python
# creating data frame where we can get the count of ratings w.r.t movies

df.groupby('title')['rating'].count().sort_values(ascending = False).head()

#this is displaying number of ratings
```

```python
#putting the mean value that we got in a data frame

ratings = pd.DataFrame(df.groupby('title')['rating'].mean())

ratings.head()
```

```python
#integrating count value table in the above dataframe
 
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())

ratings.head()
```

# Recomending Similar Movies

### Creating a Pivot Table based upon user ID and ratings given

```python
# reason for creating this pivot table is beacuse it will recomend movies according to the ratings given by each user id

moviemat = df.pivot_table(index='user_id',columns='title',values='rating')

moviemat.head()
```

```python
#sorting ratings dataframe that we created in highest to lowest format

ratings.sort_values('num of ratings',ascending=False).head(10)
```

```python
ratings.head()
```

```python
#grabbing the user ratings for two movies
# we will take any two movies and try to correlate with the pivot table that we created for recomendation process

starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()

#correlation :- statictical measure of relationship between two variables
```

```python
#we will take the above data and try to corelate with the pivot table

#for this we will use corrwith() method to get corelations between two panda series

similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
#Creating Dtaframe for shwowing correlation

corr_starwars = pd.DataFrame(similar_to_starwars,columns=['correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()

#the correlation max value will be 1....so the movie more closer to 1 or equal to 1 will likely to be our next recomendation
```

```python
#sorting the above table with correlation points from high to low by making ascending =  false

#this is beacuse the number of ratings for these movies will be once or twice

#this is based on how many users have given 

corr_starwars.sort_values('correlation',ascending=False).head(10)
```

```python
#we will filter this based upon the number of reviews

#we will discard the movies which is less than 100

corr_starwars = corr_starwars.join(ratings['num of ratings'])  #joining the number of rating columns to correlation dataframe
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('correlation',ascending=False).head()  #filtering so that number of ratings greater than 100 only is displayed in DF
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('correlation',ascending=False).head()
```
